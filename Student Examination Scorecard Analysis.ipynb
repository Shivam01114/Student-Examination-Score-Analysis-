```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Gender Distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x  = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# from the above chart we analysed that the number of females data is more than Male data
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize = (5,4))
sns.heatmap(gb , annot = True)
plt.title("Relationship between Parent Education and Student Score")
plt.show()
```

```python
# from the above chart we have concluded that the education of Parents have a good impacts on the students
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (5,4))
sns.heatmap(gb1 , annot = True)
plt.title("Relationship between Parent Martial Status and Student Score")

plt.show()
```

```python
# from the above chart we have concluded that there is no/negligible imoact on the Students Score due to there Parent Martial Status
```

```python
gb2 = df.groupby("IsFirstChild").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb2)
```

```python
plt.figure(figsize = (4,2))
sns.heatmap(gb2 , annot = True)
plt.title("Relationship between Parent Martial Status and Student Score")

plt.show()
```

```python
# from the above analysis we concluded that there is no major difference between student scores on the behalf of there first child
```

### Distribution of Ethnic Group

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A" , "group B" , "group C" , "group D" ,"group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist , labels = l , autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
gb3 = df.groupby("PracticeSport").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb3)
```

### Average Scores by PracticeSport

```python
ax = gb3.plot(kind="bar", figsize=(10,6))

plt.title("Average Scores by PracticeSport")
plt.xlabel("PracticeSport")
plt.ylabel("Average Score")
plt.legend(["Math", "Reading", "Writing"])


for container in ax.containers:
    ax.bar_label(container, fmt="%.1f")   

plt.tight_layout()
plt.show()
```
